```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification

tokenizer = AutoTokenizer.from_pretrained("google-bert/bert-base-uncased")
model = AutoModelForSequenceClassification.from_pretrained("google-bert/bert-base-uncased", 
                                                           num_labels= len(unique_bands),     
                                                           problem_type="single_label_classification")
```

```python
def tokenize(batch):
    texts = [p + tokenizer.sep_token + e for p, e in zip(batch["prompt"], batch["essay"])]
    enc   = tokenizer(texts, max_length=512, truncation=True, padding=False)
    enc["labels"] = batch["labels"]
    return enc
```

```python
from datasets import Dataset

# turn the DataFrames into Dataset objects
train_ds = Dataset.from_pandas(train_df_prediction, preserve_index=False)
test_ds  = Dataset.from_pandas(test_df_prediction,  preserve_index=False)

# now your original call works
train_ds = train_ds.map(tokenize,
                        batched=True,
                        remove_columns=train_ds.column_names)   # drop raw cols after tokenisation
test_ds  = test_ds.map(tokenize,
                       batched=True,
                       remove_columns=test_ds.column_names)

train_ds.set_format("torch")   # if you plan to pass into a Trainer
test_ds .set_format("torch")
```

```python
from evaluate import load

metric_acc = load("accuracy")
metric_f1  = load("f1")

def compute_metrics(eval_pred):
    logits, labels = eval_pred
    preds = np.argmax(logits, axis=1)
    return {
        "accuracy": metric_acc.compute(predictions=preds, references=labels)["accuracy"],
        "f1_macro": metric_f1.compute(predictions=preds, references=labels,
                                      average="macro")["f1"]
    }
```

```python
from transformers import TrainingArguments, Trainer

training_args = TrainingArguments(
    output_dir='results',        
    optim="adamw_torch",
    learning_rate=2e-5,               # learning rate
    num_train_epochs=20,              # total number of training epochs
    per_device_train_batch_size=32,  # batch size per device during training
    per_device_eval_batch_size=32,   # batch size for evaluation
    warmup_steps=500,                 # number of warmup steps for learning rate scheduler
    weight_decay=0.01,                # strength of weight decay
    report_to="none",
    logging_dir='./logs',             # directory for storing logs
    logging_steps=100,
)
```

```python
trainer = Trainer(
    model               = model,
    args                = training_args,
    train_dataset       = train_ds,
    eval_dataset        = test_ds,
    tokenizer           = tokenizer,
    compute_metrics     = compute_metrics,
)
```

```python
trainer.train()
```

```python
import numpy as np
trainer.evaluate(eval_dataset=test_ds)
```
